## 1.2) Operación básicas de vectores 

### 1.2.1) Suma 

$$ X+ Y=\begin{bmatrix}
{x_{1}}\\
{\vdots}\\
{x_{n}}
\end{bmatrix}+ 
\begin{bmatrix}
{y_{1}}\\
{\vdots}\\
{y_{n}}
\end{bmatrix}
=\begin{bmatrix}
{x_{1}+y_{1}}\\
{\vdots}\\
{x_{n}+y_{n}}
\end{bmatrix} $$

#### Propiedades 

\begin{align}
&\text{Conmutativa: }  x+y=y+x \\
&\text{Asociativa: } (x+y)+z=y+(x+z)\\
&\text{Existencia de elemento neutro: }x+0=0+x=x \\
&\text{Existencia del vector inverso en la suma: }x+(-x)=0
\end{align}

### 2.1.4) Multiplicación matriz-matriz 

$$ A B=\begin{bmatrix}
{a_{11}\quad\cdots\quad a_{1n}}\\
{\vdots \quad \vdots}\\
{a_{m1}\quad\cdots\quad a_{mn}}
\end{bmatrix}
\begin{bmatrix}
{b_{11}\quad\cdots\quad b_{1p}}\\
{\vdots \quad \vdots}\\
{b_{n1}\quad\cdots\quad b_{np}}
\end{bmatrix}=
\begin{bmatrix}
{c_{11}\quad\cdots\quad c_{1p}}\\
{\vdots \quad \vdots}\\
{c_{m1}\quad\cdots\quad c_{mp}}
\end{bmatrix}
$$

$$\LARGE A_{m\times n}B_{n\times p}=C_{m\times p}$$

$$c_{ij}=\sum_{l=1}^{n}a_{il}b_{lj}\quad \text{donde}\quad i=1,\cdots,m \quad \text{y}\quad j=1,\cdots,p $$

$$ A B=\begin{bmatrix}
{0\quad2}\\
{1\quad4}
\end{bmatrix}
\begin{bmatrix}
{1\quad 3}\\
{2\quad 1}
\end{bmatrix}=
\begin{bmatrix}
0\times 1+2\times 2 \quad 0\times 3+2\times 1\\
1\times 1+2\times 4 \quad 1\times 3+4\times 1
\end{bmatrix}=
\begin{bmatrix}
{4\quad2}\\
{9\quad7}
\end{bmatrix}
$$

Propiedades

\begin{align}
&\text{Asociativa: } (AB)C=A(BC)\\
&\text{Multipliación por un escalar: } \alpha(AB)=(\alpha A)B \\
&\text{Distributiva: }A(B+C)=AB+AC \\
&\text{Transpuesta del producto: } (AB)^{T}=B^{T}A^{T}
\end{align}

### 2.1.7) Marriz Inversa 

$$A\in \mathbb{R}^{n\times n}, \text{ si }\exists B\in \mathbb{R}^{n\times n}\\
\text{tal que } AB=BA=I_{n}\\
B=A^{-1}\\
\quad A\text{ es no singular}\Rightarrow   A \text{ es inversible}\\  \text{matriz no singular}|A| \neq 0
$$

Propiedades

\begin{align}
&a. (A^{-1})^{-1}=A\\
&b. (AB)^{-1}=B^{-1}A^{-1}\\
&c. (A^{T})^{-1}=(A^{-1})^{T}
\end{align}

### Ejemplo 19

Hallar la matriz inversa

```python
A = np.array([[1, 2, 1],
              [4, 4, 5],
              [6, 7, 7]])
```

```python
A_i = np.linalg.inv(A)
A_i 
```

```python
A@A_i,0
```
