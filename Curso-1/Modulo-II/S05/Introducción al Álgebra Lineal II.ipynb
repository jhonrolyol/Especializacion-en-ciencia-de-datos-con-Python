```python
'''
En resumen, este código implementa un algoritmo para calcular el determinante
de una matriz cuadrada utilizando la eliminación gaussiana, que consiste en 
transformar la matriz original en una matriz triangular superior y luego multiplicar
los elementos de la diagonal principal para obtener el determinante.
'''
import numpy as np
vec = []
A = np.array([[3,4,2],
              [-2,3,5],
              [4,2,6]])
for i in range(0,(len(A)-1)):
    vec.append(A[i][i])
    A[i]=(1/A[i][i])*A[i]
    for j in range((i+1),len(A)):
        if (A[j][i]*A[(i)][i])>0:
            A[j]=A[j][i]*A[(i)]-A[j]            
        else:    
            A[j]=-A[j][i]*A[(i)]+A[j]           
vec.append(A[(len(A)-1)][(len(A)-1)]) 
det=np.prod(vec)        
```

```python
def gauss(a,b):
    n = len(b)    
    for k in range(0,n-1):
        for i in range(k+1,n):
            if a[i,k] != 0.0:               
                lam = a [i,k]/a[k,k]                
                a[i,k+1:n] = a[i,k+1:n] - lam*a[k,k+1:n]                
                b[i] = b[i] - lam*b[k]                
    for k in range(n-1,-1,-1):
        b[k] = (b[k] - np.dot(a[k,k+1:n],b[k+1:n]))/a[k,k]
    
    return b
```

# 3) Valores propios y vectores propios

$$\Large{A\nu=\lambda \nu} \\ \text{donde: }\\
A \in \mathbb{R^{n\times n}}\\ 
\nu \text{ vector propio} \\
\lambda \text{ valor propio}
$$

```python
Ejemplo
```

$$ A=\begin{bmatrix}
3 & 1\\
2 & 2
\end{bmatrix}
$$

$$ \nu=\begin{bmatrix}
1 \\
-2 
\end{bmatrix}
$$

$\nu$ es un Vector propio al Valor propio $\lambda=1$ 

$$ A\nu=\begin{bmatrix}
3 & 1\\
2 & 2
\end{bmatrix}
\begin{bmatrix}
1 \\
-2 
\end{bmatrix}=
\begin{bmatrix}
1 \\
-2 
\end{bmatrix}=
1\nu
$$

También 

$$ \mu=\begin{bmatrix}
1 \\
1 
\end{bmatrix}
$$

$\mu$ es un Vector propio al Valor propio $\lambda=4$ 

$$ A\mu=\begin{bmatrix}
3 & 1\\
2 & 2
\end{bmatrix}
\begin{bmatrix}
1 \\
1 
\end{bmatrix}=
\begin{bmatrix}
4 \\
4 
\end{bmatrix}=4
\begin{bmatrix}
1 \\
1 
\end{bmatrix}=
4\mu
$$
