##### Ejemplo 35
Usando operadores de formato
<br>
d, i   Entero
<br>
f      Punto flotante como m.ddddd (primer número es el ancho del campo y el segundo número es el número de dígitos que se mostrarán después del punto decimal)
<br>
s      cadena o string

```python
precio = 24
item = "pan"
print("El %s cuesta %d centavos"%(item,precio))
print("El %+10s cuesta %5.2f centavos"%(item,precio))
print("El %+10s cuesta %10.2f centavos"%(item,precio))
```
