# Agrupando Datasets (Sentencias groupby, agg y pivot_table)

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

df
```

##### Ejemplo 51
Agrupar por columna A y luego aplicando la función sum() a los grupos resultantes para la columna C y D

```python
df.groupby("A")[["C", "D"]].sum()
```

##### Ejemplo 52
Agrupar por varias columnas (A y B) luego aplicar la función sum() a los grupos resultantes

```python
df.groupby(["A", "B"]).sum()
```

##### Ejemplo 53
Agrupar por varias columnas (A y B) luego aplicar la función mean() a los grupos resultantes

```python
df.groupby(["A", "B"]).mean()
```

##### Ejemplo 54
Agrupar por varias columnas (A y B) luego aplicar la función std() a los grupos resultantes

```python
df.groupby(["A", "B"]).std()
```

##### Ejemplo 55
Agrupar por varias columnas (A y B) luego aplicar la función count() a los grupos resultantes

```python
df.groupby(["A", "B"]).count().reset_index()
```

##### Ejemplo 56
Agrupar por la columna A y luego aplicar la función 'min', 'max', 'mean', 'median' a la columna C y 
'count', 'sum', 'std' a la columna D

```python
df_agre1 = df.groupby(['A']).agg({
    'C': ['min', 'max', 'mean', 'median'],
    'D': ['count', 'sum', 'std']
})
df_agre1
```

```python
df_agre1 = df.groupby(['A']).agg({
    'C': ['min', 'max', 'mean', 'median'],
    'D': ['count', 'sum', 'std']
}).reset_index()
df_agre1
```

##### Ejemplo 57
Filtrar solo los casos que tengan valor "foo" en la columna A.
Luego agrupar por la columna B y luego aplicar la función 'min', 'max', 'mean', 'median' a la columna C y 
'count', 'sum', 'std' a la columna D

```python
df.query('A  == "foo"')
```

```python
df_agre2 = df.query('A  == "foo"')\
.groupby(['B']).agg({
    'C': ['min', 'max', 'mean', 'median'],
    'D': ['count', 'sum', 'std']
}).reset_index()
df_agre2
```

##### Ejemplo 58

```python
df
```

```python
df_pivot = df[['A','B','C']].pivot_table(index='A',columns=['B'],aggfunc='sum')
df_pivot
```
