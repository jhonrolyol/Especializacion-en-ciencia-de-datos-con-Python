##### Ejercicio 7
Agrupa el conjunto de datos por la columna "Género" y utiliza el mean y median para obtener la edad promedio y mediana de cada grupo.

```python
datos
```

```python
datos.groupby(['Género']).agg({
    'Edad': ['mean', 'median']
})
```

##### Ejercicio 8
Utiliza el método pivot_table() para crear una tabla dinámica que muestre la cantidad de personas por Género y Ciudad.

```python
datos
```

```python
tabla = datos.pivot_table(values='ID', index='Género', columns='Ciudad', aggfunc='count')
print(tabla)
```

##### Ejercicio 9
Utiliza el método groupby() para agrupar el conjunto de datos por la columna "Ciudad" y utiliza el método sum() para obtener la suma de las edades de cada grupo.

```python
datos
```

```python
datos.groupby("Ciudad")[["Edad"]].sum()
```

##### Ejercicio 10
Utiliza el método merge() para unir el conjunto de datos "datos" con otro conjunto de datos llamado informacion_adicional en base a la columna "ID".

```python
datos
```

```python
Informacion_adicional = pd.DataFrame(
    {
        "ID": [1, 2, 3, 4, 5, 6],
        "Talla": [1.90, 1.75, 1.60, 1.69, 1.85, 1.74],
        "Peso": [85,74,65,71,82,76]
    }
)
Informacion_adicional
```

```python
pd.merge(datos, Informacion_adicional , how='outer', on='ID') 
```
