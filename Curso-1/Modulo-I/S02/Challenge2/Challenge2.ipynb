```python
# Agrupamos el conjunto de datos por la columna "Region" 
# y utilizamos "count, min, max, mean, median" para obtener 
# las estadísticas descriptivas de las variables "Units Sold, 
# Unit Price, Unit Cost, Total Revenue,Total Cost y Total Profit".
df_sales.groupby(['Region']).agg(
    {
        'Units Sold':['count','min', 'max','mean', 'median'],
        'Unit Price':['count','min', 'max','mean', 'median'],
        'Unit Cost':['count','min', 'max','mean', 'median'],
        'Total Revenue':['count','min', 'max','mean', 'median'],
        'Total Cost':['count','min', 'max','mean', 'median'],
        'Total Profit':['count','min', 'max','mean', 'median']
    }
)
```

```python
# Agrupamos el conjunto de datos por la columna "Region y Sales Channel" 
# y utilizamos "count, min, max, mean, median" para obtener 
# las estadísticas descriptivas de las variables "Units Sold, 
# Unit Price, Unit Cost, Total Revenue,Total Cost y Total Profit".
df_sales.groupby(['Region', 'Sales Channel']).agg(
    {
        'Units Sold':['count','min', 'max','mean', 'median'],
        'Unit Price':['count','min', 'max','mean', 'median'],
        'Unit Cost':['count','min', 'max','mean', 'median'],
        'Total Revenue':['count','min', 'max','mean', 'median'],
        'Total Cost':['count','min', 'max','mean', 'median'],
        'Total Profit':['count','min', 'max','mean', 'median']
    }
)
```

```python
# Seleccionar las filas de df_sales que tengan en la columna "Total Profit" 
# valores mayores o iguales que 35000
df_sales[df_sales["Total Profit"] >= 35000]
```

```python
# Seleccionemos las filas que tengan en la columna
# "Total Profit" valores mayores o iguales a 100000 y 
# en la columna "Units Sold" menores o iguales a 3000
df_sales[(df_sales["Total Profit"] >= 100000) & (df_sales["Units Sold"] <= 3000)]
```

```python
# Seleccionemos las filas que tengan en la columna
# "Item Type" el texto Vegetables o 
# en la columna "Sales Channel" el texto Offline.
    # Nota: usamos el operador "o" que se representa por "|"
df_sales[(df_sales["Item Type"] == "Vegetables") | (df_sales["Sales Channel"] == "Offline")]
```

```python
# En este caso seleccionamos las filas que tengan las columnas
# "Item Type y Sales Channel" igual a Vegetables y Offline, respectivamente.
df_sales[(df_sales["Item Type"] == "Vegetables") & (df_sales["Sales Channel"] == "Offline")]
```

```python
# Seleccionemos ahora las filas de la columna "Total Cost" 
# que tengan un costo tatal mayor a 370 000
df_sales[["Total Cost"]][df_sales["Total Cost"] > 370000]
```

```python
# Recordemos que el orden no altera el resultado en este caso
df_sales[df_sales["Total Cost"] > 370000][["Total Cost"]]
```

```python
# Recordemos que no solamente se puede colocar una columna, sino
# varias y ello va depender del interes que se tiene (en este caso usamos tres columnas).
df_sales[["Total Revenue", "Total Cost", "Total Profit"]][df_sales["Total Cost"] > 370000]
```

```python
# Seleccionar de la base de datos df_sales las filas que en la columna "Country"
# tengan los paises incluidos en la lista "["Republic of the Congo", "Belgium",
# "Norway", "Somalia", "Ghana"]"
    # Nota: Usaremos el "isin()"
df_sales[df_sales["Country"].isin(["Republic of the Congo", "Belgium", "Norway", "Somalia", "Ghana"])]
```
