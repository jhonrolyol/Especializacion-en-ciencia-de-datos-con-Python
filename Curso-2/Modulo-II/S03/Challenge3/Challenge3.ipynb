```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib
```

```python
path = "C:/Users/PC/DataAnalitica/Análisis Exploratorio/Módulo 2_Visualización de Datos con Python/sesión 3/Challenge3/"
train = path + "train.csv"
```

```python
train = pd.read_csv(train, sep = ",")
train.head()
```

```python
train.tail()
```

```python
train.info()
```

# Gráfica 1
LotShape: forma general de la propiedad
- Reg: Regular	
- IR1: Ligeramente irregular
- IR2: moderadamente irregular
- IR3: Irregular


```python
plt.figure(figsize = (12,8))
ax = sns.histplot(data = train, x='LotShape', kde=False)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}', (p.get_x() + p.get_width()/2., height),
                ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

# Gráfica 2

- SalePrice: precio de venta

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = train['SalePrice'], width=0.4, color='blue', fliersize=5)
plt.show()
```

# Gráfica 3

- LotArea: tamaño del lote en pies cuadrados
- SalePrice: precio de venta
- SaleType: tipo de venta

```python
plt.figure(figsize = (12,8))
x_axis = train['LotArea']
y_axis = train['SalePrice']
sns.scatterplot(x = x_axis, y = y_axis, s=120)
plt.show()
```

```python
plt.figure(figsize = (12,8))
x_axis = train['LotArea']
y_axis = train['SalePrice']
sns.scatterplot(x = x_axis, y = y_axis, hue=train.SaleType, s=50)
plt.show()
```

# Gráfica 4
LotShape: forma general de la propiedad
- Reg: Regular	
- IR1: Ligeramente irregular
- IR2: moderadamente irregular
- IR3: Irregular

SalePrice: precio de venta


```python
plt.figure(figsize = (12,8))
paleta= ["purple","orange","green","blue"]
sns.set_palette(paleta)
sns.catplot(x = "LotShape",
            y = "SalePrice",
            data = train,
            kind = "bar")
plt.show()
```

# Gráfica 5

* LotFrontage: Pies lineales de calle conectados a la propiedad
* LotArea: Tamaño del lote en pies cuadrados
* SalePrice - el precio de venta de la propiedad en dólares. Esta es la variable objetivo que está tratando de predecir.
* MoSold: Mes vendido


```python
plt.figure(figsize = (12,8))
data = train[['LotFrontage', 'LotArea', 'SalePrice', 'MoSold']]
sns.heatmap(data.corr(), annot=True, vmin=-1, vmax=1);
plt.show()
```
