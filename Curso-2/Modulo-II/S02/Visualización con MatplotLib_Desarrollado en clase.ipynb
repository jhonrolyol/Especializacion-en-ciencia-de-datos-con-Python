#### Interfaz Pyplot

* Permite llevar a cabo la función plot: listas, arrays
* Personalizar las líneas,colores, leyendas, etc.
* Ofrece un conjunto de tipos de graficas: hist(), scatter(), boxplot(), etc.
* Identificar los componentes:
    * Datos: Listas, ndarray, Pandas Series
    * Elementos: Titulos, nombre de los ejes, etc. Se accede a los elementos básicos con **.plt**

Caso 2
* Datos: Array
* Representación: Una función en un solo sistema de ejes

```python
x=np.linspace(-1,5,20)
```

```python
x
```

```python
def f(x):
    return np.exp(-x**2)
```

```python
# Calculando la función a los valores del objeto x
f(x)
```

#### Ejemplo
Muestre un gráfico para la función f(x). 
<br>
Título: "Grafica de la función f(x)"
<br>
X: "Eje X"
<br>
Y: "Eje Y"
<br>
Eje X: [-1, 5]
<br>
Eje Y: [-0.5, 2]
<br>
Leyenda: 'f(x)'

```python
# Elementos: .plt()
# Representando un solo gráfico

plt.figure(figsize = (10,5)) # Aumentar el tamaño de la ventana (ancho, alto) en pulgadas
plt.title("Grafica de la función f(x)")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axis((-1,5,-0.5,2))
plt.plot(x,f(x),
         marker = 'o',
         linewidth = 6, #Aumentar el tamaño de la línea
         color = 'r',
         alpha = 0.8, #Tonalidad de la línea
         markersize=15,
         linestyle = '--',
         label = 'f(x)')
plt.legend()
plt.show()
```

```python
# Otro ejemplo
npoints = 100

X = np.random.randn(npoints)
Y = -0.4 * X
plt.plot(X,Y, label="otra linea")
plt.grid()   #agrega cuadriculas al gráfico
plt.legend()
plt.show()
```

Caso 3
* Datos: Array
* Representar: Dos gráficos en un mismo sistema de Ejes

```python
x=np.linspace(-1,5,20)
x
```

```python
def fCuadratica(x):
    return x**2

def fCubica(x):
    return x**3

def fQuinta(x):
    return x**5
```

#### Ejemplo
Muestre un gráfico para la función cuadrática y cúbica en mismo sistema de ejes

```python
plt.plot(x,fCuadratica(x),
         #marker = '*',
         #color = 'b',
         #linestyle = '-',
         '*b-',
         label ='F(x)**2'
         )
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.title("Mis gráficos")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.plot(x,fCuadratica(x),
         #marker = '*',
         #color = 'b',
         #linestyle = '-',
         '*b-',
         label ='F(x)**2'
         )

plt.plot(x,fCubica(x),
         #marker = 'o'
         #color = 'k',
         #linestyle = '--',
         'ok--',
         label ='F(x)**3')
plt.legend(loc = 'best')
plt.show()
```

Para otros tipos de estilos: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

Ubicación de leyenda: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
plt.figure(figsize=(8,4))
plt.title("Mis gráficos")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.plot(x,fCuadratica(x),
         #marker = '*',
         #color = 'b',
         #linestyle = '-',
         '*b-',
         label ='F(x)**2'
         )

plt.plot(x,fCubica(x),
         #marker = 'o'
         #color = 'k',
         #linestyle = '--',
         'ok--',
         label ='F(x)**3')

plt.plot(x,fQuinta(x),
         #marker = 'o'
         #color = 'g',
         #linestyle = '--',
         'og--',
         label ='F(x)**5')

plt.legend(loc = 'best')
plt.show()
```

### 2. Histograma

Los histogramas representan en altura la acumulación de caso en intervalos de valor. Son apropiados para variables cuantitativas continuas.

```python
# Definamos la lista values
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
```

##### Ejemplo
Genere el histograma de la lista values 

```python
plt.hist(values, bins=3)
plt.show()
```

```python
# Generemos datos aleatorios normales
data_normal_1 = np.random.normal(loc = 1.0, scale = 0.5, size = 1000)

# loc: Valor "float", representa el centro de la distribución
# scale: Valor "float", representa la desviación Estandar de la distrib.
# size: un valor "int"
data_normal_1
```

```python
print(data_normal_1.shape)
print(data_normal_1.dtype)
print(data_normal_1[0:30])
```

#### Ejemplo
Graficar un histograma para data_normal_1

```python
plt.title("Histograma")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.hist(data_normal_1,
         bins = 40,
         alpha = 0.6,
         edgecolor ='k',
         label ="Histograma Data 1")
# bins: numero de Intervalos de clase
# alpha: Nitidez -[0-1]
plt.legend()
plt.show()
```

#### Ejemplo
Al histograma anterior agrege que en el eje y solo se muestren las etiquetas 0,20,40,60,80

```python
plt.title("Histograma")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.hist(data_normal_1,
         bins = 40,
         alpha = 0.6,
         edgecolor ='k',
         label ="Histograma Data 1")
# bins: numero de Intervalos de clase
# alpha: Nitidez -[0-1]

plt.yticks([0,20,40,60,80])

plt.legend()
plt.show()
```

```python
data_normal_2 = np.random.normal(loc = 1.5, scale = 0.1, size = 500)
data_normal_3 = np.random.normal(loc = 0.5, scale = 0.6, size = 500)
```

#### Ejemplo
Graficar un histograma para data_normal_2 y data_normal_3 en mismo sistema de ejes

```python
plt.figure(figsize = (12,8))
plt.title("Histograma")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.hist(data_normal_2,bins = 20,alpha = 0.5, edgecolor = 'k',label ='Hist N2')
plt.hist(data_normal_3,bins = 20,alpha = 0.5,edgecolor = 'k',label = 'Hist N3')
plt.legend()
plt.savefig("Histograma.png")
plt.show()
```

### Alternativa 2

```python
# Estructura de generación de una imagen (otra forma de entenderlo)

# Importe el submódulo matplotlib.pyplot y asígnele el nombre plt
import matplotlib.pyplot as plt

# Crea una Figure y Axes con plt.subplots
fig, ax = plt.subplots() 
# Figure object: Es un contenedor que incluye todo lo qu ves en pantalla 
# Axes object: Es la parte de la gráfica que contiene los datos, es donde dibujaremos nuestros datos para visualizarlos

# Llame a la función show para mostrar el resultado
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16, 8))   # divide el espacio gráfico en 2 filas y una columna

## Se agrega la figura 1
ax1.hist(dataset['edad'])

## Se agrega la figura 2
ax2.scatter(dataset['edad'],dataset['antiguedad'],alpha =0.6,color = 'g')
ax2.set_xlabel('edad',fontsize=10) #configura la leyenda de la grafica
ax2.set_ylabel('antiguedad',fontsize=10)
ax2.set_title('Dispersión',fontsize=20)


fig.tight_layout()
plt.show()
```

### Gráfico de 4 figuras

### 7. Gráfico Circular 

```python
data=[30,20,15,55]
equipo=["Tecnología","Administración","Ventas","Operaciones"]
```

#### Ejemplo 
Cree un gráfico circular para la lista data 

```python
fig, ax = plt.subplots()

ax.pie(data)
plt.show()
```

#### Ejemplo 
Agregar las etiquetas que se encuentran en la lista equipo.
<br>
Además cambiar a colores pasteles que se definan explicitamente.
<br>
Colocarle sombra al gráfico. 

```python
fig, ax = plt.subplots()

ax.pie(data, 
       labels = equipo,
       colors=["SteelBlue","Gold","LightCoral","LimeGreen"],
       shadow=True)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(data, 
       labels = equipo,
       colors=["SteelBlue","Gold","LightCoral","LimeGreen"],
       shadow=True,
       autopct='%1.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(data, 
       labels = equipo,
       colors=["SteelBlue","Gold","LightCoral","LimeGreen"],
       shadow=True,
       autopct='%1.1f%%',
       explode = (0.1,0,0,0))
plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(data, 
       labels = equipo,
       colors=["SteelBlue","Gold","LightCoral","LimeGreen"],
       shadow=True,
       autopct='%1.1f%%',
       explode = (0.1,0,0,0),
       pctdistance=0.7)
plt.show()
```
