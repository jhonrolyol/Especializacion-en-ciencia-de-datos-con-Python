# Gráficos usando matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/PC/DataAnalitica/Análisis Exploratorio/Módulo 2_Visualización de Datos con Python/sesión 2/Challenge2/"
train = path + "train.csv"
```

```python
train = pd.read_csv(train, sep = ",")
train.head()
```

```python
train.tail()
```

```python
train.info()
```

# GRÁFICA 1 
LotShape: forma general de la propiedad

    Reg: Regular	

    IR1: Ligeramente irregular

    IR2: moderadamente irregular

    IR3: Irregular

```python
train["LotShape"]
```

```python
plt.figure(figsize = (12,8))
plt.hist(train['LotShape'])
plt.xlabel("Forma de la propiedad")
plt.ylabel("frecuencia absoluta")
plt.title("Histograma")
```

# GRÁFICA 2


SalePrice: precio de venta

```python
train["SalePrice"]
```

```python
plt.figure(figsize = (12,8))
plt.title("SalePrice")
plt.boxplot(train["SalePrice"], patch_artist=True)
plt.show()
```

# GRÁFICA 3

LotArea: tamaño del lote en pies cuadrados

SalePrice: precio de venta

```python
fig1, ax = plt.subplots(figsize=(12, 8)) # Hoja en blanco
train.plot(
    x = "LotArea", 
    y = "SalePrice", 
    kind = "scatter",
    alpha =0.6,
    color = 'g',
    ax = ax 
)
plt.xlabel("Tamaño del lote")
plt.ylabel("Precio de venta")
plt.title("Tamaño del lote y Precio de venta")
```

# GRÁFICA 4

LotShape: forma general de la propiedad

    Reg: Regular	

    IR1: Ligeramente irregular

    IR2: moderadamente irregular

    IR3: Irregular
    
SalePrice: Precio de venta

```python
plt.figure(figsize = (12,8))
plt.bar(train['LotShape'], train['SalePrice'])
plt.xlabel("Forma de la propiedad")
plt.ylabel("Precio de venta")
plt.title("Gráficas de barras")
```

# GRÁFICA 5

LotShape: forma general de la propiedad

    Reg: Regular	

    IR1: Ligeramente irregular

    IR2: moderadamente irregular

    IR3: Irregular
    
SalePrice: Precio de venta

LotArea: tamaño del lote en pies cuadrados


```python
plt.figure(figsize=(12,10)) ## inician con el tramaño 

plt.subplot(2,2,1) ## la primera posición de subplot
plt.hist(train['LotShape'])

plt.subplot(2,2,2) ## la segunda posición de subplot
plt.scatter(train['LotArea'],train['SalePrice'],alpha =0.6,color = 'g')
plt.title("Tamaño del lote y Precio de venta",fontsize=10)
plt.xlabel("Tamaño del lote",fontsize=10)
plt.ylabel("Precio de venta",fontsize=10)

## Se agrega la figura 3
plt.subplot(2,2,3) ## la tercera posición de subplot
plt.boxplot(train['SalePrice'], patch_artist=True)

## Se agrega la figura 4
plt.subplot(2,2,4) ## la cuarta posición de subplot
plt.bar(train['LotShape'], train['SalePrice'])

plt.show()
```
